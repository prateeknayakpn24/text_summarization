```python
import xlrd
from collections import Counter 
from string import punctuation
from sklearn.feature_extraction.stop_words import ENGLISH_STOP_WORDS as stop_words
```

```python
def read_data(file):
    c=2
    script=[]
    data = xlrd.open_workbook(file)
    worksheet = data.sheet_by_name('Sheet 1')
    while(True):
        try:
            worksheet.cell(c,1).value == xlrd.empty_cell.value
            script.append(worksheet.cell(c,1).value)
        except:
            break
        c+=1
    return script
```

```python
#Separating each word as a token
def word_tokenizer(s):
    tokens = []
    for word in s.split(' '):
        tokens.append(word.strip().lower())
    return tokens
```

```python
#Separating each sentence 
def sentence_tokenizer(s):
    sentences = []
    for sentence in s.split('.'):
        sentences.append(sentence.strip())
    return sentences
```

```python
#counting the frequency of each word(excluding the stopwords).More the frequency,more the value of a word.
#The sentences having the word with high frequency will be considered important sentences.
def count_words(tokens):
    word_counts = {}
    for token in tokens:
        if token not in stop_words and token not in punctuation:
            if token not in word_counts.keys():
                word_counts[token] = 1
            else:
                word_counts[token] += 1
    return word_counts
```

```python
#Calculate the most frequent words in the paragraph.
def word_freq_distribution(word_counts):
    freq_dist = {}
    max_freq = max(word_counts.values())
    for word in word_counts.keys():  
        freq_dist[word] = (word_counts[word]/max_freq)
    return freq_dist

```

```python
#Calculating the score of sentences according to the priority of the words used most frequently.
def score_sentences(sents, freq_dist, max_len=40):
    sent_scores = {}  
    for sent in sents:
        words = sent.split(' ')
        for word in words:
            if word.lower() in freq_dist.keys():
                if len(words) < max_len:
                    if sent not in sent_scores.keys():
                        sent_scores[sent] = freq_dist[word.lower()]
                    else:
                        sent_scores[sent] += freq_dist[word.lower()]
    return sent_scores

```

```python
#Arranging the sentences having the highest scores
def summarizes(sent_scores, k):
    top_sents = Counter(sent_scores) 
    summary = ''
    scores = []
    
    top = top_sents.most_common(k)
    for t in top: 
        summary += t[0].strip()+'. '
        scores.append((t[1], t[0]))
    return summary[:-1], scores
```

```python
def final_summary(scripts):
    total_summary=[]
    for text in scripts:
        
        
        tokens = word_tokenizer(text)
        sents = sentence_tokenizer(text)
        
        word_counts = count_words(tokens)
        
        freq_dist = word_freq_distribution(word_counts)
        
        sent_scores = score_sentences(sents, freq_dist)
        
        summary, summary_sent_scores = summarizes(sent_scores, 6)
        
        total_summary.append(summary)
    return total_summary
```
